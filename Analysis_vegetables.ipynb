## Vegetable's 2019 prices analysis

Analysis of 2019 vegetable prices in Mercado Central de Buenos Aires.

Prices are in USD to exclude ARS inflation effects.

## Libraries

```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import itertools
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

%matplotlib inline
```

### Data manipulation

Import, clean and prepare vegetables and dolar datasets.

```python
# Import vegetables dataset
filename = Path(Path.cwd().joinpath('data','dataset_vegetables.csv'))
dataset = pd.read_csv(filename)

# Import dolar rate exchange dataset
dolar_filename = Path(Path.cwd().joinpath('data','Datos históricos USD_ARS.csv'))
dolar = pd.read_csv(dolar_filename, decimal= ',')
```

```python
# Vegetable dataset trasnformacion
# Data type
dataset['date'] = pd.to_datetime(dataset['date'], infer_datetime_format=True)

# Dataset ordering
dataset = dataset.sort_values(by='date')

# Quantity of data per item
quantity = dataset.groupby('item', as_index = False).count()

# Remove items that were offered less than 5 days in the year
errors = quantity.loc[quantity['date'] < 5]['item']
dataset = dataset[~dataset['item'].isin(errors)]

# Remove row with price error(20 USD)
dataset = dataset.loc[~((dataset['item']=='ALBAHACA') & (dataset['date']=='2019-01-18'))]

# Dolar Dataset

# Select columns
dolar = dolar[['Fecha','Último']]

# Data type transformation
dolar['Fecha'] = pd.to_datetime(dolar['Fecha'], infer_datetime_format = True)

# Rename columns
dolar = dolar.rename(columns={'Fecha': 'date_dolar',"Último": "exchange_rate"})

# Merge vegetables and dolar datasets
dataset = dataset.merge(dolar, left_on='date', right_on='date_dolar')
dataset = dataset.drop(['date_dolar'], axis = 1)

# Calculate vegetable price in USD
dataset['price_usd'] = dataset['average_kg_price'] / dataset['exchange_rate']
dataset.head()
```

## Plot

```python
def plot_item(item,dataset, spines = False, axis = False):
    '''
    Creates a plot on item's price per year
    Parameters
    ---------
    item (string): vegetable to plot
    dataset(dataframe): dataset with item and price_usd columns
    spines(bool): False by default, if True spines are shown
    axis(bool): False by default, if True axis are shown
    Returns
    -------
    Axes (plot): Timeseries plot with price per day
    '''
    
    df = dataset.loc[dataset['item'] == item]
    
    ax.spines["top"].set_visible(spines)    
    ax.spines["bottom"].set_visible(spines)    
    ax.spines["right"].set_visible(spines)    
    ax.spines["left"].set_visible(spines) 

    ax.axis(axis)
    
    #ax.set_ylim(0,8)
    
    ax.text(0, 0.5, item, horizontalalignment='center',
            verticalalignment='center', transform=ax.transAxes)
    
    ax.plot(df['date'],df['price_usd'], color = 'green')
    
    return ax
```

```python
# List all items for analysis
items = dataset['item'].unique()

# Plot
fig, axs = plt.subplots(nrows = len(items), ncols= 1, figsize = (10,25))

# Iterate through items
for item,ax in zip(items,axs):
    plot_item(item,dataset)
```

```python
# Save plot
fig.savefig('plots/vegetables.png')
```

### Plot the 5 chepeast vegables per month

```python
# Add month column
dataset['month'] = dataset['date'].dt.month

# Calculate mean price per item per month
data = dataset.groupby(('month','item'), as_index = False).mean().sort_values(by=['month','price_usd'])

# Rank items - chepeast first
data['ranking'] = data.groupby('month')['price_usd'].rank('first')

# Select 5 cheapest veggetable per month
small_data = data.loc[data['ranking'] <= 5 ]
```

```python
# Match vegetable icon and vegetable path
remolacha_path = 'icons/beet.png'
hinojo_path = 'icons/fennel.png'
repollo_path = 'icons/cabbage.png'
papa_path = 'icons/potato.png'
cebolla_path = 'icons/onion.png'
albahaca_path = 'icons/basil.png'
zapallo_path = 'icons/pumpkin.png'
zapallito_path = 'icons/zucchini.png'
pepino_path = 'icons/cucumber.png'
acelga_path = 'icons/chard.png'
coliflor_path = 'icons/cauliflower.png'
achicoria_path = 'icons/chicory.png'
mandioca_path = 'icons/cassava.png'
zanahoria_path = 'icons/carrot.png'
acusay_path = 'icons/akusay.png'
```

```python
# Functions
def set_path(row):
    '''
    Identifies path for each row
    
    Parameters
    ---------
    row: Row in dataset
    Returns
    -------
    path(string): Path to icon that corresponds to each row
    '''
    if row["item"] == "REMOLACHA":
        return remolacha_path
    elif row["item"] == "HINOJO":
        return hinojo_path
    elif row["item"] == "REPOLLO":
        return repollo_path
    elif row["item"] == "PAPA":
        return papa_path
    elif row["item"] == "CEBOLLA":
        return cebolla_path
    elif row["item"] == "ALBAHACA":
        return albahaca_path
    elif row["item"] == "ZAP.GRANEL":
        return zapallo_path
    elif row["item"] == "ZAPALLITO":
        return zapallito_path
    elif row["item"] == "PEPINO":
        return pepino_path
    elif row["item"] == "ACELGA":
        return acelga_path
    elif row["item"] == "COLIFLOR":
        return coliflor_path
    elif row["item"] == "ACHICORIA":
        return achicoria_path
    elif row["item"] == "MANDIOCA":
        return mandioca_path
    elif row["item"] == "ZANAHORIA":
        return zanahoria_path
    elif row["item"] == "ACUSAY":
        return acusay_path
    else:
        return "no path"
    
def getImage(path):
    '''
    Converts png icon to an object to print on top of plot
    Parameters
    ----------
    path (string): Path/ Directory where png is stored
    Returns
    ------
    Offsetimage: Object to locate on top of plot
    '''
    return OffsetImage(plt.imread(path))


```

```python
# Plot
# Create column with path to icon for each vegetable
small_data = small_data.assign(path=small_data.apply(set_path, axis=1))

# Generate y axis label
labels = ['Ene','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
labels = list(itertools.chain.from_iterable(itertools.repeat(x, 5) for x in labels))

# Define x, y variables and icons
paths = small_data.path
y = small_data.month
x = small_data.ranking

# Axes settings
fig, ax = plt.subplots(figsize = (6,8))
ax.scatter(x, y, c = 'white') 
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.axes.get_xaxis().set_visible(False)

# Iterate items and generate plot
for x0, y0, path in zip(x, y,paths):
    ab = AnnotationBbox(getImage(path), (x0, y0), frameon=False)
    ax.add_artist(ab)
    
    
# Flaticon credits
txt = 'Icons made by Freepik https://www.flaticon.com/'
fig.text(.5, .05, txt, ha='center')

# Plot settings
plt.yticks(y, labels, fontname = 'Roboto')
plt.tick_params(labelsize=15)
plt.gca().invert_yaxis()
```

```python
# Save plot
fig.savefig('plots/vegetables_icons.png')
```

### Most stable and unstable prices across the year

```python
# Aggreagate data at item level and count days the vegetable was available
dataset_days_available = dataset.groupby('item', as_index = False)['date'].count()

# Select vegetables that were available at least 200 days in the year(total market open days is 245)
dataset_days_available = dataset_days_available.loc[dataset_days_available['date'] > 200]

# Dataset with vegetables all-year availablity (Standard deviation)
stable_vegetables = dataset.merge(dataset_days_available, left_on='item', right_on='item')
stable_vegetables = stable_vegetables[['item','price_usd']].groupby('item').std().sort_values(by='price_usd').head(5)
stable_vegetables.rename(columns = {'price_usd':'std_price_usd'})
```

```python
# Vegetables with the largest price unstability (Standard deviation)
unstable_vegetables = dataset[['item','price_usd']].groupby('item').std().sort_values(by='price_usd', ascending = False).head(5)
unstable_vegetables.rename(columns = {'price_usd':'std_price_usd'})
```
